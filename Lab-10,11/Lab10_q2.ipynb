```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

data = {
    'bedrooms': [3, 4, 2, 4, 3],
    'sq_ft': [1500, 2000, 1200, 2100, 1600],
    'age': [10, 5, 20, 2, 15],
    'neighborhood': ['Urban', 'Suburban', 'Rural', 'Urban', 'Suburban'],
    'price': [300000, 450000, 200000, 480000, 320000]
}
df = pd.DataFrame(data)

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data = {
    'word_freq': [10, 1, 15, 2, 12],
    'length': [100, 50, 150, 40, 120],
    'links': [5, 0, 8, 1, 6],
    'is_spam': [1, 0, 1, 0, 1]
}
df = pd.DataFrame(data)

x = df[['word_freq', 'length', 'links']]
y = df['is_spam']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

DT = DecisionTreeClassifier()
ModelDT = DT.fit(x_train, y_train)

PredictionDT = ModelDT.predict(x_test)
print("Predictions:", PredictionDT)

testingAccDT = accuracy_score(y_test, PredictionDT) * 100
print(f"Testing Accuracy: {testingAccDT:.2f}%")
```
