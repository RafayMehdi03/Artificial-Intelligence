```python
from ortools.sat.python import cp_model

model = cp_model.CpModel()

x = model.NewIntVar(0, 20, "x")
y = model.NewIntVar(0, 20, "y")
z = model.NewIntVar(0, 20, "z")

model.Add(x + 2 * y + z <= 20)
model.Add(3 * x + y <= 18)

model.Maximize(4 * x + 2 * y + z)

solver = cp_model.CpSolver()
status = solver.Solve(model)

if status == cp_model.OPTIMAL:
    print(f"  The optimal value for x is: {solver.Value(x)}")
    print(f"  The optimal value for y is: {solver.Value(y)}")
    print(f"  The optimal value for z is: {solver.Value(z)}")
    print(f"  The maximum objective value (4x + 2y + z) is: {solver.ObjectiveValue()}")
else:
    print("Unfortunately, no optimal solution could be found given the constraints.")
```
