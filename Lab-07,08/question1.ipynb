```python
game_tree = [
  [[4, 7], [2, 5]],
  [[1, 8], [3, 6]]
]

nodes_visited = []

def run_minimax(node, level, maximizing_player):
  if level == 0 or isinstance(node, int):
    nodes_visited.append(node)
    return node

  if maximizing_player:
    highest_val = -float('inf')
    for branch in node:
      score = run_minimax(branch, level - 1, False)
      if score > highest_val:
        highest_val = score
    return highest_val
  else:
    lowest_val = float('inf')
    for branch in node:
      score = run_minimax(branch, level - 1, True)
      if score < lowest_val:
        lowest_val = score
    return lowest_val

final_score = run_minimax(game_tree, 3, True)

print(f"Root optimal value: {final_score}")
print(f"Sequence of visited nodes: {nodes_visited}")

def run_minimax_limited(node, level, maximizing_player):
  if level == 0:
    if isinstance(node, list):
      if isinstance(node[0], list):
        return node[0][0]
      else:
        return node[0]
    return node

  if isinstance(node, int):
    return node

  if maximizing_player:
    highest_val = -float('inf')
    for branch in node:
      score = run_minimax_limited(branch, level - 1, False)
      if score > highest_val:
        highest_val = score
    return highest_val
  else:
    lowest_val = float('inf')
    for branch in node:
      score = run_minimax_limited(branch, level - 1, True)
      if score < lowest_val:
        lowest_val = score
    return lowest_val

limit_score = run_minimax_limited(game_tree, 2, True)

print(f"Depth-limited result (Depth 2): {limit_score}")
```
