```python
class smartLights :
  status = False
  def __init__(self,roomName):
    self.roomName = roomName
  def lightOn(self):
    self.status = True
    print(self.roomName,"lights turned On")
  def lightOff(self):
    self.status = False
    print(self.roomName,"lights turned Off")
  def displayStatus(self):
    if(self.status == True):
      print(self.roomName,"lights are on")
    else:
      print(self.roomName,"lights are off")

s1 = smartLights("Bedroom")
s2 = smartLights("Living Room")
s2.lightOn()
s1.lightOff()
s2.displayStatus()
s1.displayStatus()

```
