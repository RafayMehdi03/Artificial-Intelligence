```python
class Vehicle:
  def __init__(self,vehicle_id,brand,rent_per_day):
    self.vehicle_id = vehicle_id
    self.brand = brand
    self.rent_per_day = rent_per_day

  def display_details(self):
    print("Vehicle Details:\nVehicle ID:",self.vehicle_id,"\nBrand:",self.brand,"\nRent:",self.rent_per_day)

  def calculate_rent(self,days):
    return(self.rent_per_day * days)

v1 = Vehicle(123,"Toyota",5000)
v2 = Vehicle(234,"Suzuki",3000)
v1.display_details()
v2.display_details()
print("Rent for Vehicle 1 for 10 days:",v1.calculate_rent(10))
print("Rent for Vehicle 2 for 15 days:",v2.calculate_rent(15))
```
