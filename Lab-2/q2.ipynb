```python
class Staff :
  def __init__(self,name,staff_id,department):
    self.name = name
    self.staff_id = staff_id
    self.department = department

class Teacher(Staff):
  def __init__(self, name, staff_id, department, courses, salary):
      super().__init__(name, staff_id, department)
      self.courses = courses
      self.salary = salary
  def teachCourse(self):
      print(self.name,"is teaching",self.courses)

class Admin(Staff):
    def __init__(self, name, staff_id, department, role,workingHours):
        super().__init__(name, staff_id, department)
        self.role = role
        self.workingHours = workingHours

    def work(self):
        print(self.name,"is handling the role of",self.role,"for",self.workingHours,"working hours")

class researchAssistants(Staff):
    def __init__(self, name, staff_id, department, stipend,research_topic):
        super().__init__(name, staff_id, department)
        self.stipend = stipend
        self.research_topic = research_topic

    def research(self):
        print(self.name,"is researching on",self.research_topic,"and has a stipend of",self.stipend)

t1 = Teacher("Ali",100,"CS","Data Structures",50000)
a1 = Admin("Moiz",866,"SE","Manager",5)
r1 = researchAssistants("Zubia",453,"CY",20000,"Aritificial Intelligence")
t1.teachCourse()
a1.work()
r1.research()
```
