```python
class Student:
    def __init__(self):
        self.__marks = 0

    def set_marks(self, marks):
        self.__marks = marks

    def get_marks(self):
        return self.__marks

    def calculate_grade(self):
        if self.__marks >= 90:
            return "A"
        elif self.__marks >= 75:
            return "B"
        elif self.__marks >= 60:
            return "C"
        else:
            return "D"


s1 = Student()
s2 = Student()
s3 = Student()
s1.set_marks(92)
s2.set_marks(68)
s3.set_marks(83)
print("Student 1:", s1.get_marks(), s1.calculate_grade())
print("Student 2:", s2.get_marks(), s2.calculate_grade())
print("Student 3:", s3.get_marks(), s3.calculate_grade())
```
