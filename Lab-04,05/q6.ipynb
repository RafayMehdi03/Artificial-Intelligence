```python
graph_astar = {
  'A': {'B': 4, 'C': 3},
  'B': {'E': 12, 'F': 5},
  'C': {'D': 7, 'E': 10},
  'D': {'E': 2},
  'E': {'G': 5},
  'F': {'G': 16},
  'G': {}
}

heuristic = {'A': 14, 'B': 12, 'C': 11, 'D': 6, 'E': 4, 'F': 11, 'G': 0}

def a_star_dynamic(graph, start, goal, heuristic):
  def search():
    frontier = [(heuristic[start], 0, start, [start])]
    visited = set()

    while frontier:
      frontier.sort(key=lambda x: x[0])
      f, g, node, path = frontier.pop(0)

      if node == goal:
        return path, g

      if node not in visited:
        visited.add(node)
        for neighbor, weight in graph[node].items():
          if neighbor not in visited:
            new_g = g + weight
            new_f = new_g + heuristic[neighbor]
            frontier.append((new_f, new_g, neighbor, path + [neighbor]))
    return None, float('inf')

  path, cost = search()
  print(path)
  print(cost)

  graph['A']['B'] = 8
  graph['B']['E'] = 7

  path, cost = search()
  print(path)
  print(cost)

a_star_dynamic(graph_astar, 'A', 'G', heuristic)
```
