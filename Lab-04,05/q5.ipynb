```python
graph_bfs = {
  'S': [('A', 3), ('B', 6), ('C', 5)],
  'A': [('D', 9), ('E', 8)],
  'B': [('F', 12), ('G', 14)],
  'C': [('H', 7)],
  'H': [('I', 5), ('J', 6)],
  'I': [('K', 1), ('L', 10), ('M', 2)],
  'D': [], 'E': [], 'F': [], 'G': [], 'J': [], 'K': [], 'L': [], 'M': []
}

def best_first_search_multi(graph, start, goals):
  full_path = [start]
  current = start
  unvisited_goals = set(goals)

  while unvisited_goals:
    frontier = [(0, current, [current])]
    visited = set()
    found_goal = False

    while frontier:
      frontier.sort(key=lambda x: x[0])
      h, node, path = frontier.pop(0)

      if node in unvisited_goals:
        unvisited_goals.remove(node)
        full_path.extend(path[1:])
        current = node
        found_goal = True
        break

      if node not in visited:
        visited.add(node)
        for neighbor, weight in graph.get(node, []):
          if neighbor not in visited:
            frontier.append((weight, neighbor, path + [neighbor]))

    if not found_goal:
      break

  print(full_path)

goals_to_visit = ['D', 'M']
best_first_search_multi(graph_bfs, 'S', goals_to_visit)
```
