```python
graph = {
  'A': ['B', 'C'],
  'B': ['D', 'E'],
  'C': ['F'],
  'D': ['G'],
  'E': [],
  'F': ['H'],
  'G': [],
  'H': []
}

def dls(graph, start, goal, depth_limit):
  visited = []

  def dfs(node, depth, path):
    if depth > depth_limit:
      return None
    visited.append(node)
    path = path + [node]
    if node == goal:
      return path
    for neighbor in graph.get(node, []):
      if neighbor not in visited:
        res = dfs(neighbor, depth + 1, path)
        if res:
          return res
    return None

  path = dfs(start, 0, [])
  print(f"Depth Limit {depth_limit}:")
  print("Nodes Visited:", visited)
  if path:
    print("Path:", path)
  else:
    print("Goal not found\n")

dls(graph, 'A', 'H', 2)
dls(graph, 'A', 'H', 3)
```
