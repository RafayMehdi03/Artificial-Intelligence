```python
graph_ucs = {
  'S': {'A': 4, 'B': 2},
  'A': {'C': 5, 'D': 10},
  'B': {'E': 3},
  'C': {'G': 4},
  'D': {'G': 1},
  'E': {'D': 4},
  'G': {}
}

def ucs(graph, start, goal):
  frontier = [(0, start, [start])]
  visited = set()

  while frontier:
    min_cost = float('inf')
    min_index = -1
    for i, (cost, node, path) in enumerate(frontier):
      if cost < min_cost:
        min_cost = cost
        min_index = i

    if min_index != -1:
      cost, node, path = frontier.pop(min_index)
    else:
      break

    if node == goal:
      print("Least Cost Path:", path)
      print("Total Cost:", cost)
      return

    if node not in visited:
      visited.add(node)
      for neighbor, weight in graph[node].items():
        if neighbor not in visited:
          new_cost = cost + weight
          new_path = path + [neighbor]
          found_in_frontier = False
          for i, (f_cost, f_node, f_path) in enumerate(frontier):
            if f_node == neighbor and f_cost > new_cost:
              frontier[i] = (new_cost, neighbor, new_path)
              found_in_frontier = True
              break
            elif f_node == neighbor and f_cost <= new_cost:
              found_in_frontier = True
              break
          if not found_in_frontier:
            frontier.append((new_cost, neighbor, new_path))

ucs(graph_ucs, 'S', 'G')
```
