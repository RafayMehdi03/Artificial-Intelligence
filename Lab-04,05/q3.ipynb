```python
def ids(graph, start, goal):
  depth = 0
  while True:
    def dfs(node, d, path):
      if d > depth:
        return None
      path = path + [node]
      if node == goal:
        return path
      for neighbor in graph.get(node, []):
        res = dfs(neighbor, d + 1, path)
        if res:
          return res
      return None

    print(f"Depth {depth}:")
    path = dfs(start, 0, [])
    if path:
      print("Final Path:", path)
      return
    depth += 1

ids(graph, 'A', 'G')
```
