```python
#same here
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination

diagnosis_net = DiscreteBayesianNetwork([
('Ailment', 'HighTemp'),
('Ailment', 'Wheeze'),
('Ailment', 'Lethargy'),
('Ailment', 'Shivering')
])

ailment_dist = TabularCPD('Ailment', 2, [[0.3], [0.7]])

temp_dist = TabularCPD('HighTemp', 2,
[[0.1, 0.5],
[0.9, 0.5]],
evidence=['Ailment'], evidence_card=[2])

wheeze_dist = TabularCPD('Wheeze', 2,
[[0.2, 0.4],
[0.8, 0.6]],
evidence=['Ailment'], evidence_card=[2])

lethargy_dist = TabularCPD('Lethargy', 2,
[[0.3, 0.7],
[0.7, 0.3]],
evidence=['Ailment'], evidence_card=[2])

shiver_dist = TabularCPD('Shivering', 2,
[[0.4, 0.6],
[0.6, 0.4]],
evidence=['Ailment'], evidence_card=[2])

diagnosis_net.add_cpds(ailment_dist, temp_dist, wheeze_dist, lethargy_dist, shiver_dist)
diagnosis_net.check_model()

solver = VariableElimination(diagnosis_net)

case_one = solver.query(
variables=['Ailment'],
evidence={'HighTemp': 1, 'Wheeze': 1}
)

print("Flu =", round(case_one.values[0], 3))
print("Cold =", round(case_one.values[1], 3))

case_two = solver.query(
variables=['Ailment'],
evidence={'HighTemp': 1, 'Wheeze': 1, 'Shivering': 1}
)

print("Flu =", round(case_two.values[0], 3))
print("Cold =", round(case_two.values[1], 3))

case_three = solver.query(
variables=['Lethargy'],
evidence={'Ailment': 0}
)

print("P(Fatigue = Yes | Flu) =", round(case_three.values[1], 3))
```
