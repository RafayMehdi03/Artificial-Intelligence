```python
import random

suits_list = ["Hearts", "Diamonds", "Clubs", "Spades"]
ranks_list = ["2", "3", "4", "5", "6", "7", "8", "9", "10",
"Jack", "Queen", "King", "Ace"]

card_deck = [(s, r) for s in suits_list for r in ranks_list]

def run_simulation(n_iters=100000):
  is_red = hearts_if_red = diamonds_if_face = spade_queen_if_face = 0
  total_red = total_face = 0

  for _ in range(n_iters):
    drawn_card = random.choice(card_deck)
    s, r = drawn_card

    if s in ["Hearts", "Diamonds"]:
        is_red += 1
        total_red += 1
        if s == "Hearts":
            hearts_if_red += 1

    if r in ["Jack", "Queen", "King"]:
        total_face += 1
        if s == "Diamonds":
            diamonds_if_face += 1
        if s == "Spades" or r == "Queen":
            spade_queen_if_face += 1

  print("P(Red):", is_red / n_iters)
  print("P(Heart | Red):", hearts_if_red / total_red)
  print("P(Diamond | Face):", diamonds_if_face / total_face)
  print("P(Spade or Queen | Face):", spade_queen_if_face / total_face)

run_simulation()
```
