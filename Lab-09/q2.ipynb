```python
#pgmpy must be installed here it isnt installed so it doesnt run.
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination

student_model = DiscreteBayesianNetwork([
('Aptitude', 'Score'),
('PrepTime', 'Score'),
('Complexity', 'Score'),
('Score', 'Success')
])

aptitude_cpd = TabularCPD('Aptitude', 2, [[0.7], [0.3]])
prep_cpd = TabularCPD('PrepTime', 2, [[0.6], [0.4]])
complexity_cpd = TabularCPD('Complexity', 2, [[0.4], [0.6]])

score_cpd = TabularCPD(
'Score', 3,
[
[0.5, 0.7, 0.2, 0.4, 0.2, 0.3, 0.05, 0.1],
[0.3, 0.2, 0.5, 0.4, 0.4, 0.4, 0.25, 0.3],
[0.2, 0.1, 0.3, 0.2, 0.4, 0.3, 0.7, 0.6]
],
evidence=['Aptitude', 'PrepTime', 'Complexity'],
evidence_card=[2, 2, 2]
)

success_cpd = TabularCPD(
'Success', 2,
[
[0.05, 0.20, 0.50],
[0.95, 0.80, 0.50]
],
evidence=['Score'],
evidence_card=[3]
)

student_model.add_cpds(aptitude_cpd, prep_cpd, complexity_cpd, score_cpd, success_cpd)

student_model.check_model()

engine = VariableElimination(student_model)

query_a = engine.query(
variables=['Success'],
evidence={'PrepTime': 0, 'Complexity': 0}
)

print("P(Success | PrepTime=Sufficient, Complexity=Hard)")
print("No:", round(query_a.values[0], 3))
print("Yes:", round(query_a.values[1], 3))

query_b = engine.query(
variables=['Aptitude'],
evidence={'Success': 1}
)

print("\nP(Aptitude | Success=Yes)")
print("High:", round(query_b.values[0], 3))
print("Low:", round(query_b.values[1], 3))
```
