```python
import numpy as np

weather_types = ["Sunny", "Cloudy", "Rainy"]

movement_rules = [
[0.6, 0.3, 0.1],
[0.3, 0.4, 0.3],
[0.2, 0.5, 0.3]
]

def run_weather_forecast(total_days):
  state = "Sunny"
  history = [state]

  for _ in range(total_days):
    if state == "Sunny":
        state = np.random.choice(weather_types, p=movement_rules[0])
    elif state == "Cloudy":
        state = np.random.choice(weather_types, p=movement_rules[1])
    else:
        state = np.random.choice(weather_types, p=movement_rules[2])

    history.append(state)

  return history

def get_rain_total(history):
  return history.count("Rainy")

forecast_sample = run_weather_forecast(10)

print("Weather Forecast for 10 Days:")
for day_idx, condition in enumerate(forecast_sample):
  print("Day", day_idx + 1, ":", condition)

sample_size = 10000
hit_target = 0

for _ in range(sample_size):
  path = run_weather_forecast(10)
if get_rain_total(path) >= 3:
  hit_target += 1

likelihood = hit_target / sample_size

print("\nProbability Analysis:")
print("At least 3 rainy days =", round(likelihood, 3))
```
