```python
option = 0
while(option != 3):
  print("1. Add Two Numbers\n2. Subtract Two Numbers\n3. Exit")
  option = int(input("Enter Your Choice: "))
  if(option == 1 or option == 2):
    x = int(input("Enter first number: "))
    y = int(input("Enter second number: "))
    if(option == 1):
      print("Result:",x+y)
    elif(option == 2):
      print("Result:",x-y)
  elif(option == 3):
      print("Exited")
  else:
      print("Invalid Option")
```
