```python
class Environment:
    def __init__(self,state = 'Heavy Traffic'):
        self.state = state

    def get_percept(self):
        return self.state

    def set_state(self,state):
        self.state = state

class ReflexiveAgent:
    def __init__(self):
        pass

    def act(self, percept):
        if percept == "Heavy Traffic":
            return "Extend Green Time"
        elif percept == "Light Traffic":
            return "Normal Green"

def run_agent(agent, environment):
    percept = environment.get_percept()
    action = agent.act(percept)
    print(f"Percept: {percept} → Action: {action}")

env = Environment()
agent = ReflexiveAgent()

run_agent(agent, env)
env.set_state('Light Traffic')
run_agent(agent,env)
```
