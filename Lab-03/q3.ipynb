```python
class Environment:
    def __init__(self, subjects):
        self.unfinished_subjects = subjects

    def study_subject(self, subject):
        if subject in self.unfinished_subjects:
            self.unfinished_subjects.remove(subject)
            return True
        return False

    def get_percept(self):
        return self.unfinished_subjects

class GoalBasedAgent:
    def __init__(self, goal):
        self.goal = goal

    def act(self, percept):
        if not percept:
            return None
        return percept[0]

def run_agent(agent, environment):
    while True:
        percept = environment.get_percept()
        subject = agent.act(percept)

        if subject is None:
            print(f"Goal Achieved: {agent.goal}")
            break

        print(f"Studying {subject}")
        environment.study_subject(subject)

subjects_to_study = ["AI", "Math", "Physics"]
env = Environment(subjects_to_study)
agent = GoalBasedAgent("All subjects completed")

run_agent(agent, env)
```
