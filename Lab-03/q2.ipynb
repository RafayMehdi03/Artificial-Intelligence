```python
import random

class Environment:
    def __init__(self):
        self.students_present = False
        self.light_status = "OFF"

    def get_percept(self):
        self.students_present = random.choice([True, False])
        return {"students": self.students_present, "lights": self.light_status}

    def set_lights(self, status):
        self.light_status = status

class ModelBasedAgent:
    def __init__(self):
        self.model = {"prev_students": None, "prev_lights": "OFF"}

    def act(self, percept):
        students = percept["students"]
        lights = percept["lights"]
        action = "No action"

        if students and lights == "OFF":
            action = "Turn lights ON"
        elif not students and lights == "ON":
            action = "Turn lights OFF"

        self.model["prev_students"] = students
        self.model["prev_lights"] = action.split()[-1] if action != "No action" else lights
        return action

def run_agent(agent, env, steps):
    for i in range(1, steps + 1):
        percept = env.get_percept()
        action = agent.act(percept)
        if "ON" in action:
          env.set_lights("ON")
        elif "OFF" in action:
          env.set_lights("OFF")
        print(f"Step {i}:")
        print(f"Percept: Students = {percept['students']}, Lights = {percept['lights']}")
        print(f"Action: {action}")
        print(f"Internal Model: {agent.model}")

env = Environment()
agent = ModelBasedAgent()
run_agent(agent, env, 8)
```
