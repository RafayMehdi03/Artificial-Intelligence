```python
class Environment:
    def __init__(self):
        self.restaurants = [
            {"name": "A", "distance": 3, "rating": 7},
            {"name": "B", "distance": 5, "rating": 9}
        ]

    def get_percept(self):
        return self.restaurants

class UtilityBasedAgent:
    def calculate_utility(self, restaurant):
        return restaurant['rating'] - restaurant['distance']

    def act(self, percept):
        best_restaurant = None
        max_utility = -float('inf')

        for res in percept:
            utility = self.calculate_utility(res)
            print(f"Resturant {res['name']} Utility = {utility}")

            if utility > max_utility:
                max_utility = utility
                best_restaurant = res['name']

        return best_restaurant

def run_agent(agent, environment):
    percept = environment.get_percept()
    selected = agent.act(percept)
    print(f"Selected Restaurant: {selected}")

env = Environment()
agent = UtilityBasedAgent()
run_agent(agent, env)
```
