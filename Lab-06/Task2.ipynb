```python
target_value = 20
beam_size = 2
def calculate_heuristic(node):
  return abs(target_value - node)

def perform_beam_search(initial_node):
  current_beam = [(initial_node, calculate_heuristic(initial_node))]
  path_tracker = {initial_node: None}
  current_level = 0
  while current_beam:
    current_beam.sort(key=lambda item: item[1])
    current_beam = current_beam[:beam_size]
    print(f"Processing Level {current_level}: Nodes in beam: {[node for node, _ in current_beam]}")
    current_level += 1
    next_beam = []
    for current_node, _ in current_beam:
      if current_node == target_value:
        final_path = []
        temp_node = current_node
        while temp_node is not None:
          final_path.append(temp_node)
          temp_node = path_tracker[temp_node]
        final_path.reverse()
        print(f"Goal {target_value} reached! Path: {final_path}")
        return

      possible_moves = [current_node + 2, current_node + 3, current_node * 2]
      for next_node in possible_moves:
        if next_node <= target_value * 2 and next_node not in path_tracker:
          next_beam.append((next_node, calculate_heuristic(next_node)))
          path_tracker[next_node] = current_node

    current_beam = next_beam
  print(f"Goal {target_value} could not be reached.")

perform_beam_search(1)
```
