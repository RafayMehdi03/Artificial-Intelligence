```python
import random

def f(x):
  return -x**2 + 6*x

def hillClimbing():
  x = random.randint(0, 6)
  print(f"Initial x: {x}")
  print(f"Initial f(x): {f(x)}")

  while True:
    neighbours = []
    if x - 1 >= 0:
      neighbours.append(x - 1)
    if x + 1 <= 6:
      neighbours.append(x + 1)

    best = x
    for neighbor in neighbours:
      if f(neighbor) > f(best):
        best = neighbor

    if best == x:
      print("No better neighbor found. Peak reached.")
      break

    x = best
    print(f"Move to x: {x}")
    print(f"f(x): {f(x)}")

  print("Final Optimal Value")
  print(f"x = {x}")
  print(f"Maximum f(x) = {f(x)}")

hillClimbing()
```
